```python
import numpy as np
import matplotlib.pyplot as plt

S = [0, 1, 2, 3]     # Estados
Q = [[1/3, 1/3, 1/3, 0], [0, 1/3, 1/3, 1/3], [1/3, 0, 1/3, 1/3], [1/2, 0, 1/2, 0]]  # Matriz de transicion
pi0  = [0, 0, 1, 0]  # Distribucion inicial


Nt = 1000

x = np.random.choice(S, p=pi0, size=Nt)  # Estado inicial


# Simulacion de la cadena de Markov
fig, ax = plt.subplots()
for t in range(30):
    for i in range(Nt):
        x[i] = np.random.choice(S, p=Q[x[i]])

ax.hist(x, color= 'black')
plt.show()
```
